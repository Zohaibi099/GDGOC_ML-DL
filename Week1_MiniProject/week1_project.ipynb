1: User Data Collector
Input handling, dictionaries, validation, formatted strings.

```python

import re

def collect_user_data():
    user_data = {}
    user_data["name"] = input("Enter your name: ")
    user_data["age"] = input("Enter your age: ")
    user_data["email"] = input("Enter your email: ")
    
    while "@" not in user_data["email"] or "." not in user_data["email"]:
        print("Invalid email! Please enter a valid email address.")
        user_data["email"] = input("Enter your email: ")

    user_data["favorite_number"] = input("Enter your favorite number: ")
    
    print(f"\nUser Details:\nName: {user_data['name']}\nAge: {user_data['age']}\nEmail: {user_data['email']}\nFavorite Number: {user_data['favorite_number']}")

collect_user_data()
    
```

2: Even or Odd
Functions, conditionals.

```python

def is_even(number):
    return number % 2 == 0

num = int(input("\nEnter a number to check even or odd: "))
print(f"The number {num} is {'even' if is_even(num) else 'odd'}.")
    
```

3: Temperature Converter
Functions, conditionals, arithmetic operations.

```python

def convert_temperature(temp, scale):
    if scale.upper() == "C":
        return (temp * 9/5) + 32
    elif scale.upper() == "F":
        return (temp - 32) * 5/9
    else:
        return "Invalid scale! Use 'C' for Celsius or 'F' for Fahrenheit."

temp_value = float(input("\nEnter temperature value: "))
temp_scale = input("Enter scale (C for Celsius, F for Fahrenheit): ")
converted_temp = convert_temperature(temp_value, temp_scale)

if isinstance(converted_temp, str):
    print(converted_temp)
else:
    print(f"Converted Temperature: {converted_temp:.2f}°{ 'F' if temp_scale.upper() == 'C' else 'C'}")
    
```

4: Finding Min & Max
Lists, functions, input handling.

```python

def find_max_min(numbers_list):
    return max(numbers_list), min(numbers_list)

numbers = [int(input(f"Enter number {i+1}: ")) for i in range(5)]
max_num, min_num = find_max_min(numbers)
print(f"Maximum: {max_num}, Minimum: {min_num}")
    
```

5: Student Data Manager
Lists, tuples, dictionaries, loops.

```python

students = []
for i in range(3):
    name = input(f"\nEnter name of student {i+1}: ")
    age = input("Enter age: ")
    grade = input("Enter grade: ")
    students.append((name, age, grade))

students_dict = {student[0]: (student[1], student[2]) for student in students}
print("\nStudent Data Dictionary:", students_dict)
    
```

6: Inventory Management System
Dictionaries, functions, conditionals.

```python

def update_inventory(inventory_dict, item, quantity):
    if item in inventory_dict:
        inventory_dict[item] = max(0, inventory_dict[item] + quantity)
    else:
        print(f"Item '{item}' not found in inventory.")
    return inventory_dict

# Initial Inventory
inventory = {
    "Apples": 10,
    "Bananas": 5,
    "Oranges": 8,
    "Milk": 2,
    "Bread": 4
}

print("\nInitial Inventory:", inventory)

# Updating Inventory
for _ in range(3):
    item_name = input("\nEnter the item name to update: ")
    item_qty = int(input("Enter quantity (use negative for removal): "))
    inventory = update_inventory(inventory, item_name, item_qty)

print("\nUpdated Inventory:", inventory)
    
```
