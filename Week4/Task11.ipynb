 Data Cleaning and Preparation Practice
Dataset: Titanic

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

1. Initial Exploration

```python
df.info()
```

```python
df.describe(include='all')
```

2. Handling Missing Data

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df['Age'])
print(df['Embarked'])
```

3. Removing Duplicates

```python
df.duplicated().sum()  # No duplicates expected
```

 4. Data Type Conversion

```python
# Convert Survived to category
df['Survived'] = df['Survived'].astype('category')
print(df['Survived'])
```

5. Renaming Columns

```python
df.rename(columns={'SibSp': 'SiblingsSpouses', 'Parch': 'ParentsChildren'}, inplace=True)
print(df)
```

6. Filtering and Sorting

```python
# Filter passengers who paid more than $100
high_fare = df[df['Fare'] > 100]
high_fare.sort_values(by='Fare', ascending=False).head()
```

7. Creating New Columns

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior'])
print(df['AgeGroup'])
```
